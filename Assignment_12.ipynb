1) In what modes should the PdfFileReader() and PdfFileWriter() File objects will be opened?
- rb and wb respectively.

2) From a PdfFileReader object, how do you get a Page object for page 5?
- by calling pdfReader.getPage(4)

3) What PdfFileReader variable stores the number of pages in the PDF document?
- numPages attribute

4) If a PdfFileReader object’s PDF is encrypted with the password swordfish, what must you do before you can obtain Page objects from it?
- Calling pdfReader.decrypt('swordfish') before reading the file.

5) What methods do you use to rotate a page?
- The pages of a PDF can also be rotated in 90-degree increments with the rotateClockwise() and rotateCounterClockwise() methods.

6) What is the difference between a Run object and a Paragraph object?
- A run object is text which has a continuous style. A new run object is generated everytime the style changes.
- Paragraph object represents the paragraphs in the document. A new para object start everytime the user presses ENTER or RETURN.

7) How do you obtain a list of Paragraph objects for a Document object that’s stored in a variable named doc?
- doc.paragraphs

8) What type of object has bold, underline, italic, strike, and outline variables?
- run object

9) What is the difference between False, True, and None for the bold variable?
-  True (the bold attribute is always enabled, no matter what other styles are applied to the run), False (the bold attribute is always disabled), or None (defaults to whatever the run’s style is set to).

10) How do you create a Document object for a new Word document?
- call docx.Document()

11) How do you add a paragraph with the text 'Hello, there!' to a Document object stored in a variable named doc?
- doc.add_paragraph('Hello, there!')

12) What integers represent the levels of headings available in Word documents?
- 0 to 4, The integer 0 makes the heading the Title style, which is used for the top of the document. Integers 1 to 4 are for various heading levels, with 1 being the main heading and 4 the lowest subheading.
