5) Create a programme that searches a folder tree for files with a certain file extension (such as .pdf or .jpg). 
Copy these files from whatever location they are in to a new folder

```python
import os,shutil

def copyExtension(folder,ext,dest):
    folder = os.path.abspath(folder)
    dest = os.path.abspath(dest)
    for dir,sub,files in os.walk(folder):
        for file in files:
            name,ext1 = os.path.splitext(file)
            if ext1==ext:
                source = folder + os.path.sep + file
                shutil.copy(source,dest)

copyExtension('ineuron','.ipynb','priti')
            
```
