```python
# 1)

def stutter(word):
    part1 = word[:2] + ".. "
    return (part1 + part1 + word + "?")

stutter('hello')
```

```python
# 2)

import math

def radians_to_degrees(rad):
    return round(rad*(180/math.pi),1)

radians_to_degrees(1)
```

```python
# 3)

def is_curzon(n):
    num = 1+ 2**n
    den = 1+2*n
    if num%den ==0:
        return True
    else:
        return False

is_curzon(14)
```

```python
# 4)
import math
def area(x):
    return round(3*math.sqrt(3)*x*x/2,1)
area(1)
```

```python
# 5)

def dec_to_binary(n):
    return bin(n)[2:]
dec_to_binary(15)
```
