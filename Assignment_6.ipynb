1. What are escape characters, and how do you use them?
- They allow you to add special characters to strings. For example, you want to add an apostrophe to 'whats up',you add a backslash followed by apostrophe (the character you want to insert) i.e. 'what\'s up'.

2. What do the escape characters n and t stand for?
- n and t add a new line and tab space respectively.

3. What is the way to include backslash characters in a string?
- by writing \\

4. The string "Howl's Moving Castle" is a correct value. Why isn't the single quote character in the word Howl's not escaped a problem?
- The enclosing quotes are double quotes so thats why it doesn't cause an issue for the single quote character within. If the enclosing quotes were single quotes, then it would be an issue.

5. How do you write a string of newlines if you don't want to use the n character?
- By using multi-line strings
