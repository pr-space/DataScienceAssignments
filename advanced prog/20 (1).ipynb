```python
def secret(s):
    s1=s.replace("."," ")[2:]
    n="<p class='str'></p> "
    n=n.replace("str",s1)
    return n
secret("p.one.two.three")
```

```python
def count_lone_ones(num):
    count=0
    lst=[int(i) for i in str(num)]
    if lst[0]==1 and lst[1]!=1:
        count+=1
    if lst[-2]!=1 and lst[-1]==1:
        count+=1
    for i in range(1,len(lst)-2):
        if lst[i]==1:
            if lst[i-1]!=1 and lst[i+1]!=1:
                count+=1
    return count
count_lone_ones(101)
```

```python
def printGrid(rows,cols):
    
    rows,cols=cols,rows
    arr=[]
    x=1
    for i in range(rows):
        col = []
        for j in range(cols):
            col.append(x)
            x+=1
        arr.append(col)
    l2 =[[row[i] for row in arr] for i in range(len(arr[0]))]
    return l2
    
printGrid(3, 6)
```
