```python
def appendZero(n1,n2):
    
    if len(n1) == len(n2):
        pass
    elif len(n1) > len(n2):
        dif = len(n1) - len(n2)
        zero = '0'*dif
        n2 = zero+n2    
    else:
        dif = len(n2) - len(n1)
        zero = '0'*dif
        n1 = zero+n1

    return n1,n2   


def meme_sum(n1,n2):   

    n1,n2 = appendZero(str(n1),str(n2))

    cnt = 0
    i = 0
    sum = 0
    sumtotal = ''
    while cnt < len(n1):
        sum = int(n1[i]) + int(n2[i])
        sumtotal = sumtotal + str(sum)
        i += 1
        sum = 0
        cnt += 1
    return int(sumtotal) 
```

```python
def isPrime(n):
    if n==1:
        return True
    else:
        for i in range(2,int(n/2)+1):
            if n%i==0:
                return False
        return True

def next_prime(x):
    if isPrime(x):
        return x
    else:
        while(isPrime(x)!=True):
            x=x+1
        return x
```

```python
def ave_spd(time1,speed1,speed2):

    uptime = time1/60
    #distance travelled uphill
    distance = speed1 * uptime
    #total distance
    totaldistance = distance*2
    #time taken to downhill
    timedown = distance/speed2
    totaltime = uptime + timedown
    totalspeed = totaldistance/totaltime
    return totalspeed
```

```python
def isPrime(n):
    if n==1:
        return True
    else:
        for i in range(2,int(n/2)+1):
            if n%i==0:
                return False
        return True
def fact(num):
    factorial=1
    for i in range(1,num + 1):
        factorial = factorial*i
    #   print(factorial)
    return factorial
    
def kempner(n):
    if isPrime(n):
        return n
    else:
        for i in range(1,n):
            if fact(i)%n == 0:
                return i
```
