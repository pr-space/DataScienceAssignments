```python
def remove_letters (l,w):
    n=[]
    for i in w:
        n.append(i)
    for item in n:
        if item in l:
            l.remove(item)
    return l
```

```python
def blocks(l):
    a=5
    b=7
    if(l==1):
        return a
    else:
        for i in range (1,l):
            a=a+b
            b=b+1
        return a
```

```python
def my_sub(a,b):
    n=[1 for i in range(a)]
    m=[1 for i in range(b)]
    for item in n:
        if item in m:
            m.remove(item)
    return sum(m)
```

```python
def add_bill(r):
    n=[]
    l =list(r.split(","))
    l=[i for i in l if i[0] == 'd']
    l=[i[1:] for i in l if i[0] == 'd']
    for i in l:
        if "k" in i:
            i = i[:-1]
            i= int(i)*1000
            n.append(i)
        else:
            n.append(int(i))
    return sum(n)
```

```python
def flip_list(l):
    y = []
    for i in l:
        if type(i)!= list:
            listA = [str(i)]
            for elem in listA:
               t = elem.split(', ')
               y.append((t))
        else:
            y.append(i[0])
    return y
```
