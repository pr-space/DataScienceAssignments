```python
def anagram(s,l):
    length = 0
    s1=''
    for i in l:
        length = length+len(i)
        if i:
            s1 +=i
    s1=s1.lower()

    s = s.replace(" ", "")
    s=s.lower()
    if sorted(list(s)) == sorted(list(s1)):
        return True
    else:
        return False
```

```python
def leaderboards(l):
    p=l
    t=[]
    j=0
    for i in p:
        x=(2*i["reputation"]) + i["score"]
        t.append(x)
        i.update(trueScore = x) 
    p = sorted(p, key = lambda i: i['trueScore'],reverse=True)
    t= sorted(t,reverse=True)
    for i in p:
        i.pop('trueScore')
        print("{0} , #trueScore = {1}".format(p[j], t[j]))
        j+=1
    
```

```python
leaderboards([
  { "name": "a", "score": 100, "reputation": 20 },
  { "name": "b", "score": 90, "reputation": 40 },
  { "name": "c", "score": 115, "reputation": 30 },
])
```

```python
def hangman(s,l):
    Temp = ""
    for letter in s:
        if letter in l:
            Temp += letter
        else:
            Temp += "-"
    return Temp
```
