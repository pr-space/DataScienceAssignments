```python
def transpose_matrix(A):
    N=len(A)
    M=len(A[0])
 
    for i in range(N):
        for j in range(i+1, M):
            A[i][j], A[j][i] = A[j][i], A[i][j]
            
    print("Transposed matrix is")
    
    for i in range(N):
        for j in range(M):
            print(A[i][j], " ", end='')
        print()
```

```python
def is_valid_hex_code(s):
    if s[0] != "#":
        return False
    else:
        s=s[1:]
        for i in s:
            if i.isalpha():
                i=i.lower()
                if ord(i)<97 or ord(i)>102:
                    return False
            elif i.isdigit():
                if int(i)<0 or int(i)>9:
                    return False
            else:
                return False
        return True
```

```python
def mark_maths(lst):
    sign=["+","-","="]
    res=0
    for exp in lst:
        temp=[]
        start = 0
        for i in range(0,len(exp)-1):
            if exp[i]== '=':
                temp.append(exp[i+1:])
            elif exp[i+1] in sign and exp[i].isdigit():
                temp.append(exp[start:i+1])
                start=i+1
        if(int(temp[0])+int(temp[1])==int(temp[2])):
            res+=1
    return (str(round((res/len(lst))*100)) +"%" ) 
```
