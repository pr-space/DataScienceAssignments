```python
def track_robot(lst):
    start=[0,0]
    for i in lst:
        way,val=i.split(" ")
        if way =="right":
            start[0] +=int(val)
        elif way =="left":
            start[0] -=int(val)
        elif way =="up":
            start[1] +=int(val)
        else:
            start[1] -=int(val)
    return (start)
track_robot([])
```

```python
def find_longest(string):
    new=string.split()
    largest=""
    for i in new:
        temp=""
        for j in i:
            if j.isalpha():
                temp+=j
        if len(temp)> len(largest):
            largest = temp

    return (largest)
    
find_longest("A thing of beauty is a joy forever.")

find_longest("Forgetfulness is by all means powerless!")
```

```python
# [very easy, very easy, easy, easy, medium, medium, hard, hard]

def interview(lst,taken_time):
    std =[5,5,10,10,15,15,20,20]
    total_time=120
    if len(lst)<len(std):
        return "disqualified"
    if taken_time>total_time:
        return "disqualified"
    for i in range(len(lst)):
        if lst[i]>std[i]:
            return "disqualified"
    return "qualified"
```
