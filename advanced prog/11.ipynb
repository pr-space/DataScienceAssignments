```python
def isPrime(num1): 
    if num1 > 1:
        for i in range(2,num1):
            if (num1%i) == 0:
                #print(num1,'is not a prime')                
                break
        else:
            return True
    else:
        return False
    
def filter_primes(lst):
    new_lst = [i for i in lst if isPrime(i)]
    return new_lst
```

```python
# 2)

def pop(lst):
    length = len(lst)//2
    fw = bw = length
    while(bw > 0):
        lst[bw] = length 
        lst[fw] = length
        bw -= 1
        fw += 1
        length -= 1
    return lst
```

```python
# 3)
def loves_me(n):
    lst =[]
    for i in range(1,n+1):
        if i % 2 == 0:
            lst.append("Loves me not")
        else:
            lst.append("Loves me")

    lst[-1] = lst[-1].upper()
    return ", ".join(lst)
```
