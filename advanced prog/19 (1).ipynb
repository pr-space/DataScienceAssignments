```python
def checker_board(n,i1,i2):
    if(i1==i2):
        return "invalid"
    else:
        matrix=[]
        for i in range(n):          
            a =[] 
            for j in range(n):    
                a.append(i1)
#                 temp=i1
#                 i1=i2
#                 i2=temp   
                i1, i2 = i2, i1
            matrix.append(a)
            if n%2==0:
                i1, i2 = i2, i1
        return matrix
checker_board(2, 7, 6)
```

```python
def almost_palindrome(s):
    rs=s[::-1]
    count=0
    for i in range(len(s)):
        if s[i]!=rs[i]:
            count+=1
    if count >2 :
        return False
    return True

    
almost_palindrome("abcdcbg")
```

```python
def prime_numbers(n):
    count=0
    for i in range(n+1):
        if i>1:
            for j in range(2,i):
                if(i % j==0):
                    break
            else:
    #             print(i)
                count+=1
    return count
                       
prime_numbers(10)
```
