```python
def only_alpha(string):
    new=""
    for i in string:
        if i.isalpha():
            new +=i
    return new
def count_same_ends(s):
    s=s.lower()
    s=s.split(" ")
    count=0
    s=[only_alpha(i) for i in s]
    for i in s:
        if i[0]==i[-1]:
            if len(i) > 1:
                count+=1
    return count
count_same_ends("Pop! goes the balloon")
```

```python
def atbash(string):
    n = ord('z') + ord('a')
    N = ord('A') + ord('Z')
    ans=''
    string=string.split(" ")
    for s in string:
        for i in s:
            if i.isalpha(): 
                if i.isupper():
                    ans+=(chr(N-ord(i)))
                else:
                    ans+=(chr(n-ord(i)))
            
            else:
                ans+=i
        ans+=" "
    ans=ans[:-1]
    return ans
atbash("apple")
```

```python
def can_see_stage(mat):
    for i in range(len(mat[0])):
        new = True
        for j in range(len(mat)-1,-1,-1):
            if new == True:
                max = mat[j][i]
                new = False
            elif (max<=mat[j][i]):
                return False
            else:
                max = mat[j][i]
            if j == 0:
                new = True

    return True
can_see_stage([
  [1, 2, 3],
  [4, 5, 6],
  [7, 8, 9]
])
```
