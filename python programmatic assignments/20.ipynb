```python
# 1)

def filter_list(lst):
    return list(filter(lambda x:type(x)==int,lst))

filter_list([1,2,3,"f"])
```

```python
# 2)

def add_indexes(lst):
    result =[]
    for i,el in enumerate(lst):
        result.append(i+el)
    return result

add_indexes([1,4,3,2,1])
```

```python
# 3)

import math
pi = math.pi
  
# Function to calculate Volume of Cone
def volume(r, h):
    return round((1 / 3) * pi * r * r * h,2)

volume(1,2)
```

```python
# 5)

def missing_no(lst):
    
    for i in range(1,11):
        if i not in lst:
            return i
        
missing_no([1,2,3,4,5,6,8,9,10])


```

```python
# 4)

def triangular_series( n ):
    j = 1
    k = 1
     
    # For each iteration increase j
    # by 1 and add it into k
    for i in range(1, n ):
        j = j + 1 # Increasing j by 1
         
        # Add value of j into k and update k
        k = k + j
    return k     
# Driven Code
n = 6
triangular_series(n)
```
