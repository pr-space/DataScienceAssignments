```python
# 1)

def equal(a,b,c):
    lst = [a,b,c]
    for i in list(set(lst)):
        if lst.count(i)!=1:
            return lst.count(i)
    return 0

equal(1,12,12)
```

```python
# 3)

def function3(lst):
    return {x:x.upper() for x in lst}

function3(["a","b"])
```

```python
# 4)

def replaceVowelsWithK(test_str, K):
 
    # string of vowels
    vowels = 'AEIOUaeiou'
 
    # iterating to check vowels in string
    for ele in vowels:
 
        # replacing vowel with the specified character
        test_str = test_str.replace(ele, K)
 
    return test_str
```

```python
# 5)

def function5(string):
    result = ""
    for i in string:
        if ord(i)%2==0:
            result += i.upper()
        else:
            result += i.lower()
    return result

function5("to be or not to be")
```
