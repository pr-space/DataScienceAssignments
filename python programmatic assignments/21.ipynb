```python
# 2)

def get_budgets(dict1):
    budget = 0
    for key in dict1:
        if key=="budget":
            budget += dict1[key]
    return budget
```

```python
# 4)

def each_year(P, r, n, t):

   
    amount = P*(pow((1+(r/n)), n*t))

    return (amount)

each_year(10000,10,.06,12)
```
