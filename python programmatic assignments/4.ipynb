```python
# Write a Python Program to Find the Factorial of a Number?

def findFactorial(p):
    if p==0:
        return 1
    elif p>0:
        n=1
        for i in range(1,p+1):
            n=n*i
        return n
    else:
        print("please use a positive whole number")

findFactorial(3)
```

```python
# Write a Python Program to Display the multiplication Table?
def showMultiplicationTable(p):
    for i in range(1,11):
        print('{} x {} = {}'.format(str(p),str(i),str(p*i)))

showMultiplicationTable(7)        
```

```python
# Write a Python Program to Print the Fibonacci sequence?

def fibonacci():
    a = int(input("How many terms do you want to print?"))
    x,y = 0,1
    if a==1:
        return x
    elif a<0:
        print("Please enter a positive number and run this program again")
    else:
        for i in range(1,a):
            print(x)
            xy = x + y
            x = y
            y = xy

fibonacci()
    
```

```python
# Write a Python Program to Check Armstrong Number?
num = int(input("Enter a number:"))
def checkArmstrong(num):
    
    if num<0:
        print("please enter a positive number")
    else:
        order = len(str(num))
        temp = num
        sum = 0
        while temp>0:
            digit = temp%10
            sum += digit**order
            temp = temp // 10
        if sum == num:
            return True
        else:
            return False

checkArmstrong(num)
```

```python
# Write a Python Program to Find Armstrong Number in an Interval?
def findArmstrong():
    a = int(input("Enter lower limit for the interval"))
    b = int(input("Enter higher limit for the interval"))
    
    if a<0 or b<0:
        print('please enter a positive number to continue')
    else:
        for i in range(a,b+1):
            if checkArmstrong(i):
                print(i)

findArmstrong()
```

```python
# Write a Python Program to Find the Sum of Natural Numbers?
def sumOfNaturals():
    num = int(input("Enter a natural number "))
    sum = 0
    if num>0:
        for i in range(1,num+1):
            sum+=i
        return sum
    else:
        print("please enter a natural number to continue")

sumOfNaturals()
```
