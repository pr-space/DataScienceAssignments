```python
# 1)

def double_char(word):
    result = ""
    for char in word:
        result += char*2
    return result
double_char("hello")
```

```python
# 2)

def reverse_bool(value):
    if value == True:
        return False
    elif value == False:
        return True
    else:
        return "Boolean expected"
    
reverse_bool("f")
```

```python
# 3)

def layers(n):
    return .5*2**n/1000

layers(1)
```

```python
# 4)

def index_caps(word):
    caps = "ABCDEFGHIJKLMOPQRSTUVWXYZ"
    result = []
    for i,char in enumerate(word):
        if char in caps:
            result.append(i)
    return result

index_caps("HeLLo")
```

```python
# 5)

def even(n):
    return [x for x in range(1,n+1) if x%2==0]

even(4)
```
