```python
# 1) Write a Python Program to Add Two Matrices? 

X = [[2,3,4],[1,3,4]]
Y = [[12,13,14],[21,23,24]]
sum = [[0,0,0],[0,0,0]]
def add_matrices(X,Y):
    
    for i in range(len(X)):
        for j in range(len(X[i])):
            sum[i][j] = X[i][j] + Y[i][j]
    return sum
add_matrices(X,Y)
```

```python
# 2) Write a Python Program to Multiply Two Matrices? 
X = [[2,3,4],[1,3,4]]
Y = [[12,13,14],[21,23,24]]
product = [[1,1,1],[1,1,1]]
def multiply_matrices(X,Y):
    
    for i in range(len(X)):
        for j in range(len(X[i])):
            product[i][j] = X[i][j] * Y[i][j]
    return product
multiply_matrices(X,Y)
```

```python
# 3) Write a Python Program to Transpose a Matrix? 
X = [[2,3,4],[1,3,4]]
result = [[0,0],[0,0],[0,0]]

def transpose_matrix(X):
    for i in range(len(X)):
        for j in range(len(X[i])):
            result[j][i] = X[i][j]
    return result

transpose_matrix(X)    
```

```python
# 4) Write a Python Program to Sort Words in Alphabetic Order? 

def sort_words():
    a = input("Enter a string: ")
    words = [word.lower() for word in a.split()]
    words.sort()
    return words
sort_words()
```

```python
# 5) Write a Python Program to Remove Punctuation From a String? Page 1 of 1
def remove_punctuation(string):
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    result = ""
    
    for char in string:
        if char not in punctuations:
            result = result + char
    return result

remove_punctuation("Hello! I am here-- Can you meet me??")
```
