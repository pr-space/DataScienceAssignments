```python
# 1)

def is_symmetrical_num(n):
  return str(n) == str(n)[::-1]
```

```python
# 2)

def multiply(str):
    prod = 1
    for num in str.split(", "):
        prod *= int(num)
    return prod

multiply("11, 22, 10")
```

```python
# 3)

def square_digits(n):
    n1 = str(n)
    lst = [str(int(x)**2) for x in n1]
    return ("").join(lst)

square_digits(123)
```

```python
# 4)
def sortify(l):
    l.sort()
    return list(set(l))

sortify([5,5,1,1,1,1,2,3])
```

```python
# 5)

def find_mean(n):
    l = len(str(n))
    num = n
    sum = 0
    while num!=0:
        sum+= num%10
        num = num//10
    return sum/l

find_mean(12345)
```
