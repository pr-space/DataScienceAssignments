```python
# 1)
def filter_strings(lst):
    return list(filter(lambda x: type(x)==int,lst))

filter_strings([1,2,"a"])
```

```python
# 2)

def string_reversed(str):
    return str[::-1].swapcase()

string_reversed("Hello! World! rEvERseE")
```

```python
# 3)

def list_split(lst):
    first,*middle,last = lst
    print(first)
    print(middle)
    print(last)

list_split([1,2,3,4,5])
```

```python
# 4)

def factorial(n):
    if n==0:
        return 1
    else:
        return n*factorial(n-1)
    
factorial(0)
```

```python
# 5)

def moving(lst,n):
    result1 = []
    result2= []
    for el in lst:
        if (el==n):
            result2.append(el)
        else:
            result1.append(el)
    return result1 + result2
moving([1,5,5,5,4,5,6,7,5],5)    
```
