```python
# Write a Python program to find sum of elements in list?

def find_sum(l):
    sum = 0
    for i in l:
        if type(i)==int:
            sum += i
    return sum

find_sum([1,12,'h',True,3])
```

```python
# Write a Python program to  Multiply all numbers in the list?
def find_product(l):
    prod = 1
    for i in l:
        if type(i)==int:
            prod *= i
    return prod

find_product([1,12,'h',True,3])
```

```python
# Write a Python program to find smallest number in a list?

def find_smallest(l):
    temp = []
    for i in l:
        if type(i)==int:
            temp.append(i)
    smallest = temp[0]
    for i in temp:
        if i<=smallest:
            smallest = i
    return smallest

find_smallest([111,12,'h',True,30])
```

```python
# Write a Python program to find largest number in a list?
def find_largest(l):
    temp = []
    for i in l:
        if type(i)==int:
            temp.append(i)
    largest = temp[0]
    for i in temp:
        if i>=largest:
            largest = i
    return largest

find_largest([111,122,'h',True,30])
```

```python
# Write a Python program to find second largest number in a list?
def find_largest(l):
    m = l.copy()
    temp = []
    for i in m:
        if type(i)==int:
            temp.append(i)
    temp.sort()
    return temp[-2]

find_largest([11,2,'h',True,30])
```

```python
# Write a Python program to find N largest elements from a list?
def filter_list(l):
    temp = []
    for i in l:
        if type(i)==int:
            temp.append(i)
    return temp

def find_n_largest(l,n):
    temp = filter_list(l)
    temp.sort(reverse=True)
    return temp[:n]
    
find_n_largest([111,2,3,4,55,-10],3)
```

```python
# Write a Python program to print even numbers in a list?
def print_even(l):
    temp = filter_list(l)
    return list(filter(lambda x:x%2==0,temp))
    
print_even([2,3,1,2,3,0,4,11])
```

```python
# Write a Python program to print odd numbers in a List?
def print_odd(l):
    temp = filter_list(l)
    return list(filter(lambda x:x%2!=0,temp))
    
print_odd([2,3,1,2,3,0,4,11])
```

```python
# Write a Python program to Remove empty List from List?
def remove_empty_lists(l):
    return list(filter(lambda x:x!=[],l))

remove_empty_lists([[],2,3,'f',[]])
```

```python
# Write a Python program to Cloning or Copying a list?
def clone_list(l):
    return l.copy()

clone_list([1,2,3])
```

```python
# Write a Python program to Count occurrences of an element in a list?
def count(l,e):
    elements_in_list = list(filter(lambda x:x==e,l))
    return len(elements_in_list)
count([1,12,12,1,1,10],12)
```
