```python
# 2)

def simon_says(l1,l2):
    if l1[:-1]==l2[1:]:
        return True
    return False

simon_says([1, 2], [5, 5])
```

```python
# 3)

def secret_society(lst):
    result = []
    for name in lst:
        result.append(name[:1])
    return ("").join(sorted(result))

secret_society(["hello","abby","new"])
```

```python
# 4)

def is_isogram(word):
    result = []
    for char in word.lower():
        if char in result:
            return False
        else:
            result.append(char)
    return True

is_isogram("Consecutive")
```

```python
# 5)

def is_sorted(word):
    sorted_word = ("").join(sorted(word))
    if word == sorted_word:
        return True
    return False

is_sorted("14123")
```
