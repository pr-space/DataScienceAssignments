```python
# Write a Python Program to find sum of array?

def find_sum(arr):
    sum = 0
    for i in arr:
        sum += i
    return sum

find_sum([1,2,3])
```

```python
# Write a Python Program to find largest element in an array?
def find_largest(arr):
    largest = arr[0]
    for i in arr:
        if i>largest:
            largest = i
    return largest
find_largest([11,112,3])
```

```python
# Write a Python Program for array rotation?

def rotate_array(arr,rot):
    temp = []
    length = len(arr)
    if rot>length:
        rot = rot%length
    for i in range(rot,length):
        temp.append(arr[i])
    
    for i in range(rot):
        temp.append(arr[i])
    return temp

rotate_array([1,2,3,4,5,6,7,8],17)
```

```python
# Write a Python Program to Split the array and add the first part to the end?
def split_arr(arr):
    length = len(arr)
    split = length//2
    temp = []
    for i in range(split,length):
        temp.append(arr[i])
    for i in range(split):
        temp.append(arr[i])
    return temp

split_arr([1,2,3,4,5,6,7])
```

```python
# Write a Python Program to check if given array is Monotonic?

def is_it_monotonic(arr):
    n = len(arr)
    return all(arr[i]<=arr[i+1] for i in range(n-1) or arr[i]>=arr[i+1] for i in range(n-1))


is_it_monotonic([61,62,63,64])      
```
