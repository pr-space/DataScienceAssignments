```python
# Question 1:
# Please write a program using generator to print the numbers which can be divisible by 5 and 7
# between 0 and n in comma separated form while n is input by console.
# Example:
# If the following n is given as input to the program:
# 100
# Then, the output of the program should be:
# 0,35,70

def gen1(n):
    def gen2(n):
        for i in range(n+1):
            if i%7==0 and i%5==0:
                yield i
    values = []
    for i in gen2(n):
        values.append(str(i))
    
    return (",").join(values)

gen1(100)
```

```python

# Question 2:
# Please write a program using generator to print the even numbers between 0 and n in comma separated form 
# while n is input by console.
# Example:
# If the following n is given as input to the program:
# 10
# Then, the output of the program should be:
# 0,2,4,6,8,10

def gen1():
    n = int(input("Enter a number: "))
    def gen2(n):
        for i in range(n+1):
            if i%2==0:
                yield i
    values = []
    for i in gen2(n):
        values.append(str(i))
    
    return (",").join(values)

gen1()
```

```python

# Question 3:
# The Fibonacci Sequence is computed based on the following formula:
# f(n)=0 if n=0
# f(n)=1 if n=1
# f(n)=f(n-1)+f(n-2) if n>1
# Please write a program using list comprehension to print
# the Fibonacci Sequence in comma separated form with a given n input by console.
# Example:
# If the following n is given as input to the program:
# 7

# Then, the output of the program should be:
# 0,1,1,2,3,5,8,13

def fibonacci(n):
    values = []
    for i in range(n+1):
        if i==0:
            values=[0]
        elif i==1:
            values=[0,1]
        elif i>1 :
            values.append(values[-1]+values[-2])
    
    return ",".join(map(str,values))

fibonacci(7)
```

```python

# Question 4:
# Assuming that we have some email addresses in the 
# "username@companyname.com" format, please write 
# program to print the user name of a given email address. Both user names and company names are composed of letters only.
# Example:
# If the following email address is given as input to the program:
# john@google.com
# Then, the output of the program should be:
# john

def get_username(email):
    return email.split("@")[0]

get_username("username@companyname.com")
```

```python

# Question 5:
# Define a class named Shape and its subclass Square.
# The Square class has an init function which takes a length as argument.
# Both classes have a area function which can print the area of the shape where Shape's area is 0 by default.

class Shape():
    def __init__(self):
        pass
    
    def area(self):
        return 0
    
class Square(Shape):
    def __init__(self,l):
        Shape.__init__(self)
        self.l = l
        
    def area(self):
        return self.l*self.l

line = Shape()
line.area()

sq = Square(6)
sq.area()
```
