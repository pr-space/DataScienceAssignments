```python
# Write a Python program to find words which are greater than given length k?

def words_greater(k):
    string = input("Enter a string: ")
    words = string.split()
    result = []
    for word in words:
        if len(word)>k:
            result.append(word)
    return result
words_greater(2)
            
```

```python
# Write a Python program for removing i-th character from a string?

def remove_i_th(string,i):
    return string[:i-1] + string[i:]

remove_i_th('Hello',3)
```

```python
# Write a Python program to split and join a string?
def split_string(string):
    list_string = string.split(' ')
    return list_string

def join_string(list_string):
    string = '-'.join(list_string)
    return string

string = 'I am doing assignments'

list_string = split_string(string)
print("After Splitting: ",list_string)

res_string = join_string(list_string)
print("After joining: ",res_string)

```

```python
# Write a Python to check if a given string is binary string or not?

def check_binary(string) :
    '''returns True if a given "string" is binary string'''
    p = set(string)
 
    s = {'0', '1'}

    if s == p or p == {'0'} or p == {'1'}:
        return True
    else :
        return False
    
check_binary("10101")
```

```python
# Write a Python program to find uncommon words from two Strings?

def find_unique(s1,s2):
    words1 = [word.lower() for word in s1.split()]
    words2 = [word.lower() for word in s2.split()]
    result = []
    for word in words1:
        if word not in words2:
            result.append(word)
    return result

find_unique('hello my name is','my name')
```

```python
# Write a Python to find all duplicate characters in string?
def find_duplicates(string):
    duplicates = []
    for char in string:
        if string.count(char)>1:
            duplicates.append(char)
    return set(duplicates)      
find_duplicates('assess')
```

```python
# Write a Python Program to check if a string contains any special character?

import re

def check(string):
 
    regex = re.compile('[@_!#$%^&*()<>?/\|}{~:]')
     
    if(regex.search(string) == None):
        return True
         
    else:
        return False
        
check("Hello!")
```
