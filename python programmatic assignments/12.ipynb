```python
# 1) Write a Python program to Extract Unique values dictionary values? 

def extractUnique(dict1):
    '''takes in a dictionary dict1 and returns Unique values dictionary values'''
    result = []
    for val in dict1.values():
        for e in val:
            result.append(e)
    return list(set(result))

extractUnique({1:[1,2,3,4,5],2:[2,3,44,55,66]})
```

```python
# 2) Write a Python program to find the sum of all items in a dictionary? 
def returnSum(myDict):
    ''' takes in a dictionary myDict and returns the sum of all its values'''
    list = []
    for i in myDict:
        list.append(myDict[i])
    final = sum(list)
     
    return final

returnSum({1:200,2:200})
```

```python
# 6) Write a Python program to check order of character in string using OrderedDict()?
 
def checkOrder(string, pattern): 
    dic = OrderedDict.fromkeys(string) 
    ptr = 0
    for key,value in dic.items(): 
        if (key == pattern[ptr]): 
            ptr = ptr + 1
        if (ptr == (len(pattern))): 
            return 'True'
    return 'False'

string = 'Study tonight'
pattern = 'stu'
print (checkOrder(string,pattern))

string2= 'Welcome'
pattern2= 'cm'
print (checkOrder(string2,pattern2)) 
```
