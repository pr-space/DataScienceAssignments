```python

# 1. What is the result of the code, and why?
def func(a, b=6, c=8):
    print(a, b, c)
func(1, 2)
# 1 2 8
# On line 6, the default values of 6 and 8 are assigned to b and c. When c is not mentioned, the defualt value is used whereas b=2 on line 6 so, default value is not used
```

```python
# 2. What is the result of this code, and why?
def func(a, b, c=5):
    print(a, b, c)
func(1, c=3, b=2)
# 1 2 3
# values assigned on line 4 override the default values assigned to b and c on line 2
```

```python
# 3. How about this code: what is its result, and why?
def func(a, *pargs):
    print(a, pargs)
    print(type(pargs))
func(1, 2, 3)
# 1 (2,3)
# 2,3 are non keyword arguments. tuple 
```

```python
# 4. What does this code print, and why?
def func(a, **kargs):
    print(a, kargs)
func(a=1, c=3, b=2)
# 1 {'c':3,'b':2}
# a is 1 where c and b are keyword arguements kwargs
# One can think of the kwargs as being a dictionary that maps each keyword to the value that we pass alongside it. 
```
