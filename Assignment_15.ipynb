7) Write a generator, genPrimes, that returns the sequence of prime numbers on successive calls to its next() method: 2, 3, 5, 7, 11, ...



```python
def genPrimes(n):
    
    def isPrime(x):
        if x==2 or x==3:
            return True
        else:
            for i in range(2,x):
                if x%i==0:
                    return False
            else:
                return True
    
    for i in range(2,n):
        if isPrime(i):
            yield i
```

```python
p=genPrimes(100)
```
