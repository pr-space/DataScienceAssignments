- Functions reduce the need for duplicate code. This makes programs shorter, easier to read and easier to update.
Reusability is another reason.Once a function is defined, it can used over and over again.
Another reason is abstraction. You dont need to know what is inside a function to use it. You only need the name of the function and the arguments to run a function.
