Q5) What is the purpose of the with/as argument, and what other statement is it like? Page 1 of 1
> The with/as statement is used in exception handling to write cleaner and more readable code.

Example:
```
# case 1
file = open('file_path', 'w')
file.write('hello world !')
file.close()
  
# case 2
file = open('file_path', 'w')
try:
    file.write('hello world')
finally:
    file.close()
 

# case 3: using with statement
with open('file_path', 'w') as file:
    file.write('hello world !')
```
There is no need to close the file in case 3 and note how much more readabe case 3 looks compared to case 1 and 2
