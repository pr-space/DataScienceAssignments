Q2) What is the concept of delegation? 
> When a class delegates one of its method to another unrelated class to finish the task, its called delegation.


Q3) What is the concept of composition? 
>  When a class delegates one of its method to another unrelated class to finish the task, its composition. The classes here have a "has a" relationship. Example: For example, your Horse class can be composed by another object of type Tail. Composition allows you to express that relationship by saying a Horse has a Tail.
