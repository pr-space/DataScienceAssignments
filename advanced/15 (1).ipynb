What is monkey patching in Python?<br>
> Dynamic modifications made at runtime to classes or modules.

What is the difference between a shallow copy and deep copy?<br>
>A shallow copy constructs a new compound object and then (to the extent possible) inserts references into it to the objects found in the original. A deep copy constructs a new compound object and then, recursively, inserts copies into it of the objects found in the original.

What is generator comprehension?<br>
>generator comprehension is a construct which you can use to create a new generator from an existing one.
```
>>> (x**2 for x in xrange(1,11))
<generator object at 0x7f0a79273488>
```
